```python
import pandas as pd
```

# Tabla KPI's

En este notebook se elaboró una tabla resumen que se usara para realizar el calculo de los KPI's del analisis.

```python
hechos = pd.read_csv('hechos.csv')
```

```python
hechos
```

## Calculo de Poblacion anual

```python
pob = pd.read_excel('Poblacion por comuna 2015a2022.xlsx')
pob.rename(columns={'Año/Comuna':'AÑO'}, inplace=True)
pob
```

## Calculo de victimas por semestre

```python
victimas_año = hechos.groupby(['AÑO','MES']).agg({'N_VICTIMAS':'sum'}).reset_index()
victimas_año
```

```python
victimas_sem1 = victimas_año[victimas_año['MES']<=6].groupby(['AÑO']).agg({'N_VICTIMAS':'sum'}).reset_index()
victimas_sem1
```

```python
victimas_sem2 = victimas_año[victimas_año['MES']>6].groupby(['AÑO']).agg({'N_VICTIMAS':'sum'}).reset_index()
victimas_sem2
```

```python
victimas_año = victimas_sem1.merge(victimas_sem2,suffixes=['_1','_2'], on = 'AÑO')
victimas_año
```

```python
victimas_melted = pd.melt(victimas_año[['AÑO','N_VICTIMAS_1','N_VICTIMAS_2']], id_vars=['AÑO'], value_vars=['N_VICTIMAS_1', 'N_VICTIMAS_2'], 
                          var_name='Tipo', value_name='N_VICTIMAS', ignore_index=False)

victimas_melted['Tipo'] = victimas_melted['Tipo'].str.extract('(\d)')
victimas_melted['SEMESTRE'] = victimas_melted['AÑO'].astype(str) + '-' + victimas_melted['Tipo']
victimas_semestre = victimas_melted.drop(columns=['AÑO', 'Tipo'])
victimas_semestre = victimas_semestre[['SEMESTRE', 'N_VICTIMAS']]
victimas_semestre=victimas_semestre.sort_values('SEMESTRE')
victimas_semestre
```

```python
victimas_semestre = victimas_semestre.join(pob[['Total']])
victimas_semestre.rename(columns={'Total':'TOTAL'},inplace=True)
victimas_semestre
```

## Calculo de victimas en MOTO

```python
victimas_moto = hechos[hechos['VICTIMA']=='MOTO'].groupby(['AÑO']).agg({'N_VICTIMAS':'sum'}).reset_index()
victimas_moto
```

## Calculo de victimas PEATON

```python
victimas_peaton = hechos[hechos['VICTIMA']=='PEATON'].groupby(['AÑO']).agg({'N_VICTIMAS':'sum'}).reset_index()
victimas_peaton
```

## Calculo de siniestros en INTERSECCION

```python
victimas_inter = hechos[hechos['INTERSECCION']==1].groupby(['AÑO']).agg({'ID':'count'}).reset_index()
victimas_inter
```

## Union de resultados

```python
victimas_sum = victimas_moto.merge(victimas_peaton,on='AÑO',suffixes=('_MOTO','_PEATON'))
victimas_sum = victimas_sum.merge(victimas_inter,on='AÑO')
victimas_sum.rename(columns={'ID':'N_INTERSECCION','Total':'POBLACION'},inplace=True)
victimas_sum
```

## Exportar archivos

```python
victimas_sum.to_csv('tabla_kpis_anual.csv', index=False)
```

```python
victimas_semestre.to_csv('tabla_kpis_semestral.csv', index=False)
```
